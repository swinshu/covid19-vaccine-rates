```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# read in data
df = pd.read_csv("data/merged.csv")
```

```python
x = np.log(df["2020"])
y = df["people_vaccinated_per_hundred"]
plt.scatter(x, y)

plt.xlabel("log GDP in USD")
plt.ylabel("People Vaccinated per 100")
plt.title("GDP vs. Vaccinations")
```
